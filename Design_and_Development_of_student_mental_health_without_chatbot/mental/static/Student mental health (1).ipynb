```python
# Define the scoring function considering all columns
def mental_health_score(row):
    score = 0
    
    # High severity indicators
    if row['Thoughts that you would be better off dead or of hurting yourself in some way'] == 'Yes':
        score += 3
    if row['Feeling down, depressed, or hopeless '] == 'Yes':
        score += 3
    if row['Little interest or pleasure in doing things '] == 'Yes':
        score += 3
    
    # Medium severity indicators
    if row['Are you having proper sleep or not minimum 6hrs daily'] == 'No':
        score += 2
    if row['Feeling tired or having little energy'] == 'Yes':
        score += 2
    if row['Feeling bad about yourself or that you are a failure or not have let yourself or your family down'] == 'Yes':
        score += 2
    if row['Did you feel trouble on concentrating on things, such as reading the newspaper or studying or reading books or while watching television and etc...'] == 'Yes':
        score += 2
    if row['Moving or speaking so slowly that other people could have noticed Or being so restless that you have been moving around a lot more than usual']== 'Yes':
        score += 2
    
    # Lifestyle and behavior
    if row['Do you consume substances such as alcohol, tobacco, or drugs?  '] == 'Yes':
        score += 1
    if row['How many hours do you spend studying each day(Self Studying)?'] < 1:
        score += 1
    if row['How many hour do you spend electronic gadgets (e.g. mobile phone, computer, laptop, PSP, PS4.0, Wii, etc.) '] > 4:
        score += 1
    if row['How many hours do you spend on social media per day?'] > 4:
        score += 1
    if row['  How many hours of physical activity or exercise do you engage in weekly?  '] < 1:
        score += 1
    
    # Demographic factors
    if row['Did you live with your family or not'] == 'No':
        score += 1
    if row['Do you have part-time or full-time job? '] == 'Yes':
        score += 1
    if row['Your Last Year Percentage'] < 50:
        score += 1  # Lower performance may be indicative of stress

    return score

# Apply the function to calculate the total mental health score
df['Mental Health Score'] = df.apply(mental_health_score, axis=1)

# Categorize based on the score
def categorize_by_score(score):
    if score >= 15:
        return 'Severely Affected'
    elif score >= 8:
        return 'Moderately Affected'
    else:
        return 'Healthy'

# Apply the function to categorize based on the score
df['Mental Health Category'] = df['Mental Health Score'].apply(categorize_by_score)

# Check the distribution of the new labels
df['Mental Health Category'].value_counts()
```
